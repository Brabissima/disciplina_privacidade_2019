```python
def distance(reg1, reg2):
    count = 0
    for i in range(1, reg1.shape[0]):
        if(reg1[i]!=reg2[i]):
            count = count + 1
    return count
```

```python
def ordenarDist(df):
    distances = []
    for i in range(0, df.shape[0]):
        for j in range(i+1,df.shape[0]-1):
            dis = distance(df.loc[[i]],df.loc[[j]])
            distances.append([dis, i, j])
    distances_Ord = sorted(distances)
    return distances_Ord
```
