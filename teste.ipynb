```python
import pandas as pd
import numpy as np
import csv
```

```python
salario_2 = pd.read_csv('salario.csv')
```

# União dos atributos de localidade

```python
df = salario_2.copy() 
local = []
cidades = salario_2.cidade
estados = salario_2.estado
paises = salario_2.pais

for i in range(0,len(cidades)):
    local.append(str('{}, {}, {}'.format(cidades[i], estados[i], paises[i])))
```

```python
pd.DataFrame(local)
df = df.drop("cidade",axis=1)
df = df.drop("estado",axis=1)
df = df.drop("pais",axis=1)

date = []
datas = salario_2.data

for i in range(0,len(datas)):
    date.append(datas[i])

df = df.drop("data",axis=1)

df['local'] = local
df['data'] = date

df
```

# Função de Precisão

```python
def precisao(N, D, h, HGV): #N: número de atributos, D: número de registros, h: lista com altura da árvore de cada atributo, HGV: lista com alturas máximas das hierarquias dos atributos
    somatorio = 0
    for i in range(0,N):
        somatorio += h[i]/HGV[i]
    #como o somatório interno não está sendo utilizado como índice, basta multiplicar o somatório por D
    resultado = 1 - ((D*somatório)/(D*N))
    return resultado
```
