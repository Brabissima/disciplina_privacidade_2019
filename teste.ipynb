# Função de Precisão

```python
def precisao(N, D, h, HGV): #N: número de atributos, D: número de registros, h: lista com altura da árvore de cada atributo, HGV: lista com alturas máximas das hierarquias dos atributos
    somatorio = 0
    for i in range(0,N):
        somatorio += h[i]/HGV[i]
    #como o somatório interno não está sendo utilizado como índice, basta multiplicar o somatório por D
    resultado = 1 - ((D*somatório)/(D*N))
    return resultado
```

# Distância

```python
def dist(reg1, reg2):
    genero1 = np.array(reg1["genero"])
    local1 = np.array(reg1["local"].str.split(", ", n=2, expand=True))[0]
    data1 = np.array(reg1["data"].str.split("/", n=2, expand=True))[0]
    genero2 = np.array(reg2["genero"])
    local2 = np.array(reg2["local"].str.split(", ", n=2, expand=True))[0]
    data2 = np.array(reg2["data"].str.split("/", n=2, expand=True))[0]
    distValor = 0
    distTexto = ""
    if(genero1[0] != genero2[0]):
        distValor = distValor + 1
        distTexto = distTexto + "g"
    if(data1[2] != data2[2]):
        distValor = distValor + 3
        distTexto = distTexto + "a"
    else:
        if(data1[1] != data2[1]):
            distValor = distValor + 2
            distTexto = distTexto + "m"
        else:
            if(data1[0] != data2[0]):
                distValor = distValor + 1
                distTexto = distTexto + "d"
    if(local1[2] != local2[2]):
        distValor = distValor + 3
        distTexto = distTexto + "p"
    else:
        if(local1[1] != local2[1]):
            distValor = distValor + 2
            distTexto = distTexto + "e"
        else:
            if(local1[0] != local2[0]):
                distValor = distValor + 1
                distTexto = distTexto + "c"
    return distValor, distTexto
```
