```python
import pandas as pd
import numpy as np
import csv
```

```python
read = pd.read_csv('salario.csv')
salario_2 = read.drop('salario', axis=1)
salario_2
```

```python
df = salario_2.copy() 
local = []
cidades = salario_2.cidade
estados = salario_2.estado
paises = salario_2.pais

for i in range(0,len(cidades)):
    local.append(str('{}, {}, {}'.format(cidades[i], estados[i], paises[i])))
```

```python
pd.DataFrame(local)
df = df.drop("cidade",axis=1)
df = df.drop("estado",axis=1)
df = df.drop("pais",axis=1)

date = []
datas = salario_2.data

for i in range(0,len(datas)):
    date.append(datas[i])

df = df.drop("data",axis=1)

df['local'] = local
df['data'] = date

df
```
